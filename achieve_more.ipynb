```python
import os
from dotenv import load_dotenv
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import WebBaseLoader
from langchain_google_genai.embeddings import GoogleGenerativeAIEmbeddings
from scipy.spatial.distance import cosine
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
load_dotenv()
embeddings = GoogleGenerativeAIEmbeddings(google_api_key=os.getenv("GEMINI_API_KEY"), model='gemini-embedding-001')
```

```python
text1 = 'The cat sat on the mat'
text2 = 'I am going to build an AI agent.'
vect1 = embeddings.embed_query(text1)
vect2 = embeddings.embed_query(text2)

print(f'{1 - cosine(vect1,vect2):.4f}')
```

```python
docs = ['i am nitesh', 'i live in biratangar', 'i love coding in python', 'i love to wake up 6 am mornig', 'i extermly love to do workout', 'i hate king']
```

```python
db = FAISS.from_texts(docs, embeddings)
```

```python
simlarity = db.similarity_search('hate', k=3)
simlarity
```
